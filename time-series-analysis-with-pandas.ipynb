Time series analysis usually focuses on datasets where the data points are organized according to time. This can include examples like stock prices tracked over multiple days or weather data recorded over various time intervals.
So we are adding BoardingTime column to perfom "Time series analysis".

```python
import numpy as np

start_time = pd.Timestamp('1912-04-10 09:00:00')
time_intervals = pd.to_timedelta(np.arange(len(df)) * 10, unit='min')
df['BoardingTime'] = start_time + time_intervals

df.to_csv('updated_titanic.csv', index=False)
print(df[['PassengerId', 'BoardingTime']].head())
```

Resampling the data

```python
resampled_data = df.resample('h', on='BoardingTime').size()
print(resampled_data.head())
```

Rolling Statistics

```python

df['RollingFareMean'] = df['Fare'].rolling(window=5).mean()

print(df[['PassengerId', 'Fare', 'RollingFareMean']].head(10))
```

Lagging the Data

```python

df['Fare_Lagged'] = df['Fare'].shift(1)

print(df[['PassengerId', 'Fare', 'Fare_Lagged']].head(10))
```
